```python
import numpy as np
import h5py

datafile = h5py.File('SCNeuronModelCompetition.mat')
movie = datafile.get('trainingmovie_mini') # movie for training
frhist = datafile.get('FRhist_tr') # firing rate histograms
```

```python
print(movie.shape)
print(frhist.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
idx = 30 # movie chunk index
view_movie = movie[idx].reshape(-1, 96, 128).astype('uint8')
view_hist = frhist[idx]
cnk = 5
for m in range(cnk, 150, cnk):
    img = view_movie[m]
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.subplot(1, 2, 2)
    plt.plot(view_hist[m])
#     plt.plot(np.mean(frhist[m-cnk:m, :, n], axis=0))
    plt.show()
    print(np.argmax(view_hist[m])) # neuron that fired the most
    print(np.where(view_hist[m] > 0)) # list of neurons that fired at all
```
